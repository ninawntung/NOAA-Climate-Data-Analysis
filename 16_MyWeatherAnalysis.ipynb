### Tenperature chart - short timeframe (Monthly Basis)

```python
import csv
import matplotlib.pyplot as plt
from datetime import datetime

#open file
filename = 'sitka_weather_07-2018_simple.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    # Get the headers
    for index, column_header in enumerate(header_row):
        print(index, column_header)

    # Get high temperatures, dates
    highs = []
    dates = []

    for row in reader:
        high = int(row[5])
        current_date = datetime.strptime(row[2], '%Y-%m-%d')
        highs.append(high)
        dates.append(current_date)

# Plot high temperatures
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(16,10))
ax.plot(dates, highs, c='red')

#Format plot
ax.set_title("Daily high temperatures, July 2018\nSitka, Alaska", fontsize=24)
ax.set_xlabel('Date',fontsize=16)
ax.set_ylabel("Temperature(F)", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)

fig.autofmt_xdate()

plt.show()


```

### Error Checking - Ignoring Missing Data

```python
import csv
from datetime import datetime
from matplotlib import pyplot as plt

#open file
filename = 'death_valley_2018_simple.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    # Get the index and headers
    for index, column_header in enumerate(header_row):
        print(index, column_header)

    # Get the dates, high temperatures and low temperatures for line chart
    dates = []
    highs = []
    lows = []

    for row in reader:
        
        date = datetime.strptime(row[2], "%Y-%m-%d")    
        try:
            high = int(row[4])
            low = int(row[5])
        except ValueError:
            print(f"Missing data for {date}")
        else:
            dates.append(date)
            highs.append(high)
            lows.append(low)

# Plot chart
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(16,10))
ax.plot(dates, highs, c='red')
ax.plot(dates, lows, c='blue')
ax.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1)

# Format chart
ax.set_title("Daily high and low temperature - 2018\nDeath Valley, CA", fontsize=24)
ax.set_xlabel("Date", fontsize=16)
fig.autofmt_xdate()
ax.set_ylabel("Temperature(F)", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)

plt.show()

    
```

### Automatic Indexes

```python
import csv
import matplotlib.pyplot as plt
from datetime import datetime

#open file
filename = 'sitka_weather_2018_simple.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    
    # Get the headers
    for index, column_header in enumerate(header_row):
        print(index, column_header)

    # Get high temperatures, dates
    highs = []
    lows = []
    dates = []

    date_index = header_row.index('TMAX')
    high_index = header_row.index('TMIN')
    low_index = header_row.index('DATE')

    for row in reader:
        high = int(row[date_index])
        low = int(row[high_index])
        current_date = datetime.strptime(row[low_index], '%Y-%m-%d')

        highs.append(high)
        lows.append(low)
        dates.append(current_date)

# Plot high temperatures
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(16,10))
ax.plot(dates, highs, c='red')
ax.plot(dates, lows, c='blue')
ax.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1)

#Format plot
ax.set_title("Daily high and low temperatures - 2018\nSitka, Alaska", fontsize=24)
ax.set_xlabel('Date',fontsize=16)
ax.set_ylabel("Temperature(F)", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)

fig.autofmt_xdate()

plt.show()
```

# Sitka & Death Valley Rainfall

```python
import csv
from datetime import datetime
from matplotlib import pyplot as plt

filename = 'sitka_weather_2018_simple.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)

    for index, column_header in enumerate(header_row):
        print(index, column_header) 

    # Get the dates and rainfalls data
    dates = []
    rainfalls = []

    for row in reader:
        date = datetime.strptime(row[2], "%Y-%m-%d")
        rainfall = float(row[3])

        dates.append(date)
        rainfalls.append(rainfall)

# Plot line chart
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(16,10))
ax.plot(dates, rainfalls, c='blue')

# Format chart
ax.set_title("Daily Rainfall - 2018\nSitka, Rainforest", fontsize=24)
ax.set_xlabel("Date", fontsize=14)
ax.set_ylabel("Rainfall (mm)")
ax.tick_params(axis='both', which='major', labelsize=14)

```

# Sitka & Death Valley Temperature Range Comparison

```python
import csv
import matplotlib.pyplot as plt
from datetime import datetime


def get_data_plot(filename, highs, lows, dates, high_index, low_index, date_index):
    #open file
    with open(filename) as f:
        reader = csv.reader(f)
        header_row = next(reader)

        for row in reader:
            try:
                high = int(row[high_index])
                low = int(row[low_index])
                current_date = datetime.strptime(row[date_index], '%Y-%m-%d')

            except ValueError:
                print(f"Missing data for {current_date}")

            else:
                highs.append(high)
                lows.append(low)
                dates.append(current_date)



sitka_dates = []
sitka_highs = []
sitka_lows = []
get_data_plot(filename='sitka_weather_2018_simple.csv', highs=sitka_highs, lows= sitka_lows, 
                dates=sitka_dates, high_index=5,low_index=6, date_index=2)

death_dates = []
death_highs = []
death_lows = []
get_data_plot(filename='death_valley_2018_simple.csv', highs=death_highs, lows= death_lows, 
                dates=death_dates, high_index=4,low_index=5, date_index=2)


# Plot temperatures
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(16,10))

#Sitka
ax.plot(sitka_dates, sitka_highs, c='red')
ax.plot(sitka_dates, sitka_lows, c='blue')
ax.fill_between(sitka_dates, sitka_highs, sitka_lows, facecolor='blue', alpha=0.1)

#Death Valley
ax.plot(death_dates, death_highs, c='red', alpha=0.5)
ax.plot(death_dates, death_lows, c='blue', alpha=0.5)
ax.fill_between(death_dates, death_highs, death_lows, facecolor='blue', alpha=0.05)

fig.autofmt_xdate()

# Format chart
ax.set_title("Sitka (Rainforest) & Death Valley(Desert)\nTemperature Range Comparison", fontsize=24)
ax.set_xlabel("Date", fontsize=14)
ax.set_ylabel("Temperature(F)")
ax.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```
